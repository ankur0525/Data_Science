```python
####there are some values with Nan, so we are going to use 2 different approach and check
#1. drop Nan valsues
#2. convert Nan values to 0
```

```python
##### drop Nan Values
anime_data_series.dropna(subset=['start_year', 'end_year'])
```

```python
anime_data_series['span'] = anime_data_series[anime_data_series['end_year'] != '']['end_year'].astype(int) - \
                            anime_data_series[anime_data_series['start_year'] != '']['start_year'].astype(int)
anime_data_series.tail(10)
```

```python
anime_data_series['end_year'] = anime_data_series['end_year'].fillna(0).astype(int)
anime_data_series['start_year'] = anime_data_series['start_year'].fillna(0).astype(int)
```

```python
anime_data_series['span'] = anime_data_series[anime_data_series['end_year'] != '']['end_year'] - \
                            anime_data_series[anime_data_series['start_year'] != '']['start_year']
anime_data_series.tail(10)
```

```python
anime_data_series_sorted_duration = anime_data_series.sort_values(by ='span', ascending = False)

plt.figure(figsize=(10,3))
plt.title('Top 10 Longest Anime')
plt.ylabel('Anime')
sns.barplot(x= 'span', y= 'Title', data=anime_data_series_sorted_duration[:10], palette = "dark")
```
